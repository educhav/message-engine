```python
import json
import pandas as pd

data = {
    "name": [],
    "message": []
}
filter_phrases = ["left the group.", "to the group.",
                  "from the group.", "added a participant.",
                 "set his own nickname", "set the nickname for",
                  "This poll is no longer available.", "in a game.",
                 "https://www", "started a call.", "named the group",
                 "cleared the nickname for", "http"]
format_tuples = [("â", "'"), ("â", "'")]

def match(message):
    for phrase in filter_phrases:
        if phrase in message:
            return False
    return True

def format_message(message):
    for f_tuple in format_tuples:
        if f_tuple[0] in message:
            message.replace(f_tuple[0], f_tuple[1])
    return message
    
    
with open("../messages/message_full.json") as file:
    for line in json.load(file)["messages"]:
        if match(line["content"]):
            data["name"].append(format_message(line["sender_name"]))
            data["message"].append(format_message(line["content"]))
 
df = pd.DataFrame(data)
df.to_csv("messages.csv", index=False)
df.sample(30)
```
